```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("istherecorrelation.csv",
                sep=';',
                decimal=',',
                index_col="Year")
df.head()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["WO [x1000]"], df["NL Beer consumption [x1000 hectoliter]"], s=80, c="royalblue", alpha=0.7, edgecolors="k")

for year, x, y in zip(df.index, df["WO [x1000]"], df["NL Beer consumption [x1000 hectoliter]"]):
    if year == 2011:
        plt.text(x, y, str(year), fontsize=10, ha="left", va="bottom")
    elif year == 2012:
        plt.text(x, y, str(year), fontsize=10, ha="left", va="top")
    else:
        plt.text(x, y, str(year), fontsize=10, ha="right", va="bottom")

plt.xlabel("WO graduates (people)")
plt.ylabel("Beer consumption (1000 hectoliters)")
plt.title("Scatter plot: WO vs Beer consumption (NL)")
plt.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.savefig("plot.png", dpi=300)
plt.show()
```
